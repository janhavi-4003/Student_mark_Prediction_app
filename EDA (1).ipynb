```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
# Load the cleaned dataset
ds = pd.read_csv('new_cleanesd_data.csv')
ds
```

# Frequency Distriubution (Barplots)

```python

# Count plots
plt.figure(figsize=(15,5))
for i, col in enumerate(['Gender', 'EthnicGroup', 'ParentEduc', 'TestPrep']):
    plt.subplot(1, 4, i+1)
    sns.countplot(x=ds[col])
    plt.xticks(rotation=45)
    plt.title(col)
plt.tight_layout()
plt.show()
```

# Distribution of Scores(Histograms)

```python
ds[['MathScore', 'ReadingScore', 'WritingScore']].hist(bins=30, figsize=(10,6))
plt.suptitle("Score Distributions")
plt.show()
```

# Boxplots to Detect Outliers

```python
plt.figure(figsize=(12,4))
for i, col in enumerate(['MathScore', 'ReadingScore', 'WritingScore']):
    plt.subplot(1, 3, i+1)
    sns.boxplot(y=ds[col])
    plt.title(f"{col} - Boxplot")
plt.tight_layout()
plt.show()
```

# Correlation Heatmap

```python
sns.heatmap(ds[['MathScore', 'ReadingScore', 'WritingScore']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation between scores")
plt.show()
```
